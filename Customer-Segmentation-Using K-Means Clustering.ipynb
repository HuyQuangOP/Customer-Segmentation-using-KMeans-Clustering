##  4. RFM Analysis

* Definition:
    RFM analysis is a widely used customer analysis method in the field of marketing and customer relationship management. RFM helps businesses better understand how they interact with each individual customer and from there classify customers into groups based on three main factors:

* **Recency (R)**: Recency measures how recent a customer's last purchase or interaction with the business was. Customers with recent activity are considered more engaged and likely to be repeat or loyal customers.

* **Frequency (F)**: Frequency measures how often a customer has made a purchase or interacted with the business within a given period. High-frequency customers are those who make regular purchases or engage with the brand consistently.

* **Monetary (M)**: Monetary value represents the total revenue generated by a customer's transactions over a specific period. High-value customers are those who spend more money and contribute significantly to the business's bottom line.

### Build features from sales data

```python
df['created_time'] = pd.to_datetime(df.created_at)
df['created_date'] = pd.to_datetime(df.created_at).dt.date
df['customer_id_len'] = df.customer_id.astype(str).str.len()
```

```python
df.customer_id_len.value_counts()
```

```python
recency_df = (
    df
    .groupby(by='customer_id')['created_time'].agg('max')
    .reset_index()
    .rename(columns = {'created_time': 'last_purchase_date'})
)
recent_date = recency_df['last_purchase_date'].max()
recency_df['recency'] = recency_df.last_purchase_date.apply(lambda x: (recent_date - x).days)
recency_df
```

```python
frequency_df = (
    df
    .drop_duplicates(subset=['order_code'], keep='last')
    .groupby(by=['customer_id'])['created_time'].agg('count')
    .reset_index()
    .rename(columns = {'created_time': 'frequency'})
)
frequency_df
```

```python
monetary_df  = (
    df
    .drop_duplicates(keep='last')
    .groupby(by=['customer_id'])['amount'].agg('sum')
    .reset_index()
    .rename(columns = {'amount': 'monetary'})
)
monetary_df
```

```python
rfm_df = recency_df.loc[: , ['customer_id', 'recency']].merge(frequency_df, how = 'left', on='customer_id')
rfm_df = rfm_df.merge(monetary_df, how = 'left', on='customer_id')
rfm_df
```

```python
rfm_df.describe().T
```

```python
plt.figure(figsize = (24,16))

plt.subplot(2,3,1)
sns.distplot(rfm_df['recency'])
plt.title('Recency Distribution', fontsize = 16)
plt.xlabel('Recency', fontsize = 14)
plt.ylabel('Density', fontsize = 14)

plt.subplot(2,3,2)
sns.distplot(rfm_df['frequency'])
plt.title('Frequency Distribution', fontsize = 16)
plt.xlabel('Frequency', fontsize = 14)
plt.ylabel('Density', fontsize = 14)

plt.subplot(2,3,3)
sns.distplot(rfm_df['monetary'])
plt.title('Monetary Distribution', fontsize = 16)
plt.xlabel('Monetary', fontsize = 14)
plt.ylabel('Density', fontsize = 14)

plt.show()
```

High-spending, frequent customers skew data. Ranking within categories based on data size (ranking) provides a score for each customer.

### Classified according to ranking formula:

```python
def calculate_rfm_scores(dataframe):
    dataframe['r_rank'] = dataframe['recency'].rank(ascending=False)
    dataframe['f_rank'] = dataframe['frequency'].rank(ascending=False)
    dataframe['m_rank'] = dataframe['monetary'].rank(ascending=False)
    
    max_r_rank = dataframe['r_rank'].max()
    max_f_rank = dataframe['f_rank'].max()
    max_m_rank = dataframe['m_rank'].max()
    
    dataframe['r_score'] = 100 * (dataframe['r_rank'] / max_r_rank).round(2)
    dataframe['f_score'] = 100 * (dataframe['f_rank'] / max_f_rank).round(2)
    dataframe['m_score'] = 100 * (dataframe['m_rank'] / max_m_rank).round(2)
    
    dataframe = dataframe.loc[:, ['r_score', 'f_score', 'm_score']]
    return dataframe
```

```python
data = calculate_rfm_scores(rfm_df)
rfm_df
```

## Segmentation based on original fomula of RFM score 

rfm score = 0.05 * (0.15 x Recency score + 0.28 x Frequency score + 0.57 x Monetary score)

Rating Customer based upon the RFM score:
* rfm score >4.5 : Top Customer
* 4.5 > rfm score > 4 : High Value Customer
* 4>rfm score >3 : Medium value customer
* 3>rfm score>1.6 : Low-value customer
* rfm score<1.6 :Lost Customer

```python
rfm_df['rfm_score'] = 0.05 * (0.15 * rfm_df['r_score'] + 0.28 * rfm_df['f_score'] + 0.57 * rfm_df['m_score'])
```

```python
def classify_customer_group(x):
    if x < 1.6:
        return 'Lost Customer'
    elif 1.6 <= x < 3:
        return 'Low value Customer'
    elif 3 <= x < 4:
        return 'Medium value Customer'
    elif 4 <= x < 4.5:
        return 'High value Customer'
    else:
        return 'Top Customer'
```

```python
rfm_df['rfm_cluster'] = rfm_df['rfm_score'].apply(classify_customer_group)
```

```python
rfm_df['rfm_cluster'].value_counts().plot(kind='pie', autopct='%1.0f%%')
```

## Segmentation based on K-means Clustering

### Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

### Shihoutte Score

```python
from sklearn.metrics import silhouette_score

for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    labels = kmeans.labels_
    score = silhouette_score(data, labels)
    print(f"For {k} clusters, silhouette score is {score:.4f}")
```

Now, I know the optimal number for clusters is **5**. It's time for model training

### Model trainning

```python
import plotly.graph_objs as go
```

```python
data = rfm_df.loc[:, ['r_score', 'f_score', 'm_score']]
model = KMeans(n_clusters = 5, init = "k-means++", max_iter = 300, n_init = 10, random_state = 0)
y_clusters = model.fit_predict(data)
data['cluster'] = y_clusters
```

```python
Scene = dict(xaxis = dict(title  = 'r_score -->'),yaxis = dict(title  = 'f_score--->'),zaxis = dict(title  = 'm_score-->'))

labels = model.labels_
trace = go.Scatter3d(x=data.loc[:, 'r_score'], y=data.loc[:, 'f_score'], z=data.loc[:, 'm_score'], mode='markers',marker=dict(color = labels, size= 10, line=dict(color= 'black',width = 10)))
layout = go.Layout(margin=dict(l=0,r=0),scene = Scene,height = 800,width = 800)
data = [trace]
fig = go.Figure(data = data, layout = layout)
fig.show()
```
